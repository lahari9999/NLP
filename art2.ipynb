```python
f2 = 'virus.txt'
```

```python
f2 = open('virus.txt','rt')
```

```python
allwords = f2.read()
```

```python
print(allwords)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens1 = tokenizer.tokenize(allwords.lower())
print(tokens1)
```

```python
len(tokens1)
```

```python
from nltk.probability import FreqDist
fdist3 = FreqDist()
for word3 in tokens1:
    fdist3[word3]+=1
fdist3
```

```python
fdist3_top25 = fdist3.most_common(25)
fdist3_top25
```

```python
from nltk.tokenize import blankline_tokenize
A_blank = blankline_tokenize(allwords)
len(A_blank)
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
len(stopwords.words('english'))
```

```python
stoptokens1 =[token for token in tokens1 if token not in stopwords.words('english')]
```

```python
print(stoptokens1)
```

```python
len(stoptokens1)
```

```python
from nltk.probability import FreqDist
fdist3 = FreqDist()
for word3 in stoptokens1:
    fdist3[word3]+=1
fdist3
```

```python
second = fdist3.most_common(25)
```

```python
second
```

```python
import pandas as pd
df = pd.DataFrame(second,columns = ['Word','Count'])
df.plot.bar(x='Word',y='Count',title = 'Most common Frequency Counts')
```
